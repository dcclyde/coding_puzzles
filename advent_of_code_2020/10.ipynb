```python
dat.sort()
min(dat)
```

```python
ones = 0
threes = 1
for k in range(len(dat) - 1):
    dx = dat[k+1] - dat[k]
    if dx == 1:
        ones += 1
    elif dx == 3:
        threes += 1

print(ones, threes, ones * threes)
```

```python
dp = [0 for x in range(len(dat))]

# dp[k] = ways to get from dat[0] to dat[k] inclusive.
dp[0] = 1
for k in range(1, len(dat)):
    offset = 1
    while k-offset >= 0 and dat[k] - dat[k-offset] <= 3:
        dp[k] += dp[k-offset]
        offset += 1

print(dp[-1])
```
